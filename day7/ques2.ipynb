```python
# 1) DataFrame using 2D List

import pandas as pd

data = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]

df = pd.DataFrame(data)

print(df)
```

```python
# 2) DataFrame using Dictionary

import pandas as pd

data = {
    "Name": ["Vikas", "Rahul", "Aman"],
    "City": ["Jaipur", "Kota", "Ajmer"],
    "Marks": [90, 85, 88]
}

df = pd.DataFrame(data)

print(df)
```

```python
# 3) DataFrame using List of Lists

import pandas as pd

data = [
    ["Virat", 18],
    ["Rohit", 45],
    ["Dhoni", 7]
]

df = pd.DataFrame(data, columns=["Player", "Jersey No"])

print(df)
```

```python
# 4) DataFrame using List of Tuples

import pandas as pd

data = [
    ("Maths", 95),
    ("Physics", 91),
    ("Chemistry", 89)
]

df = pd.DataFrame(data, columns=["Subject", "Marks"])

print(df)
```

```python
# 5) DataFrame using List of Dictionaries

import pandas as pd

data = [
    {"Name": "Vikas", "Age": 21},
    {"Name": "Rahul", "Age": 22},
    {"Name": "Aman", "Age": 20}
]

df = pd.DataFrame(data)

print(df)
```
