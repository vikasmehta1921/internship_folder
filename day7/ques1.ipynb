```python
import pandas as pd
data = {
    "maths":90,
    "chemistry":80,
    "physics":98,
    "dsa": 87
}
df = pd.Series(data)
print(df)
print(df.iloc[0]) # iloc is use to access 0 index position
print(df['physics'])
print(df[1:3])
```

```python
import pandas as pd 

data = [14,28,20,'01']

index = ['Jaipur','Baran','Kota','Ajmer']

df = pd.Series(data, index)

print(df)

print(df.iloc[1])      # Position based
print(df.iloc[0])

print(df.loc['Jaipur':'Kota'])   # Label based slicing
print(df[1:3])
print(df.loc['Jaipur':'Kota'])   # Label based slicing
print(df[1:3])
```
