```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 1. Replace NaN with 0 and interchange rows and columns

arr = np.array([
    [6, -8, 73, -110],
    [np.nan, -8, 0, 94]
])

print("Original Array:")
print(arr)

# Replace NaN with 0
arr = np.nan_to_num(arr, nan=0)

print("\nArray after replacing NaN with 0:")
print(arr)
```

```python
# Interchange rows and columns
transpose_array = arr.T

print("\nTranspose Array:")
print(transpose_array)
```

```python
# 2. Move axes of 3D array

arr3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print("\nOriginal 3D Array:")
print(arr3d)

moved_array = np.moveaxis(arr3d, [0, 1, 2], [2, 0, 1])

print("\nArray after moving axes:")
print(moved_array)
```

```python
# 3. Replace NaN values with average of columns

arr_nan = np.array([
    [10, 20, np.nan],
    [40, np.nan, 60],
    [70, 80, 90]
])

print("\nOriginal Array with NaN:")
print(arr_nan)

# Column mean
col_mean = np.nanmean(arr_nan, axis=0)

# Find NaN indices
indices = np.where(np.isnan(arr_nan))

# Replace NaN with column mean
arr_nan[indices] = np.take(col_mean, indices[1])

print("\nArray after replacing NaN with column average:")
print(arr_nan)
```

```python
# 4. Replace negative values with 0

negative_array = np.array([10, -5, 20, -8, 30, -2])

print("\nOriginal Array:")
print(negative_array)

negative_array[negative_array < 0] = 0

print("\nArray after replacing negative values with 0:")
print(negative_array)
```

```python
# 5. Average, Mean, Median, Mode of two 2D arrays

arr1 = np.array([
    [10, 20],
    [30, 40]
])

arr2 = np.array([
    [5, 15],
    [25, 35]
])
```

```python
average = (arr1 + arr2) / 2

print("\nAverage of Arrays:")
print(average)
```

```python
# Mean
mean_value = np.mean(average)

print("\nMean:")
print(mean_value)
```

```python
# Median
median_value = np.median(average)

print("\nMedian:")
print(median_value)
```

```python
# Mode
mode_value = stats.mode(average, axis=None)

print("\nMode:")
print(mode_value.mode)
```

```python

# 6. Solve equation using linalg and inverse matrix

# Coefficient matrix
A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])
```

```python
# Constant matrix
B = np.array([9, -6, 17])
```

```python
# Inverse matrix
A_inverse = np.linalg.inv(A)
```

```python
# Solution
solution = np.dot(A_inverse, B)

print("\nSolution of equations:")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])
```
