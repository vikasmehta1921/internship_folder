```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aman', 'Riya', 'Karan', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4, 5],
    'Marks': [85, 90, 88, 95]
})

print("DF1")
print(df1)

print("\nDF2")
print(df2)
```

```python
left_join = pd.merge(df1, df2, on='ID', how='left')

print("\nLeft Join")
print(left_join)
```

```python
right_join = pd.merge(df1, df2, on='ID', how='right')

print("\nRight Join")
print(right_join)
```

```python
# Set ID as index
df1_index = df1.set_index('ID')
df2_index = df2.set_index('ID')

# Join using index
index_join = df1_index.join(df2_index)

print("\nIndex Based Join")
print(index_join)
```

```python
# | Feature       | `pd.merge()`     | `df.join()`     |
# | ------------- | ---------------- | --------------- |
# | Works on      | Columns          | Index           |
# | Flexibility   | More flexible    | Simpler         |
# | Common Use    | SQL-style joins  | Index alignment |
# | Default Join  | Inner            | Left            |
# | Multiple Keys | Supported easily | Less convenient |
```
