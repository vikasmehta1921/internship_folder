```python
import pandas as pd
```

```python
# First DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Riya']
})

# Second DataFrame
df2 = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Karan', 'Neha']
})

# Third DataFrame
df3 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'City': ['Delhi', 'Mumbai', 'Pune', 'Jaipur']
})

print("DataFrame 1")
print(df1)

print("\nDataFrame 2")
print(df2)

print("\nDataFrame 3")
print(df3)
```

```python
# Concatenate df1 and df2 vertically
concat_df = pd.concat([df1, df2], axis=0)

print("Concatenated DataFrame")
print(concat_df)
```

```python
# Concatenate df1 and df2 horizontally
concat_df = pd.concat([df1, df3], axis=1)

print("Concatenated DataFrame")
print(concat_df)
```

```python
concat_df = pd.concat([df1, df2], axis=0, ignore_index=True)

print(concat_df)
```

```python
merged_df = pd.merge(concat_df, df3, on='ID')

print("Merged DataFrame")
print(merged_df)
```

```python
pd.merge(concat_df, df3, on='ID', how='inner')

#Returns only matching rows.
```

```python
pd.merge(concat_df, df3, on='ID', how='left')

#Keeps all rows from left DataFrame.
```

```python
pd.merge(concat_df, df3, on='ID', how='right')

#Keeps all rows from right DataFrame.
```

```python
pd.merge(concat_df, df3, on='ID', how='outer')

#Keeps all rows from both DataFrames.
```
